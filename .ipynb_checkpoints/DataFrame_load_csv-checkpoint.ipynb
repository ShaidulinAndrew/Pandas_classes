### __Загрузка данных из CSV файла в объект DataFrame__

```python
import pandas as pd
```

```python
# считываем содержимое файла в объект DataFrame
df = pd.read_csv('./goog.csv')
df
```

```python
df.Date
# pandas не знает что первый столбец дата,определяются как str
```

```python
type(df.Date[0])
```

```python
# Параметр parse_dates функции pd.read_csv() указывает pandas как нужно преобразовать данные непосредственно в пандосовкий объект
df = pd.read_csv('./goog.csv', parse_dates = ['Date'])
df
```

```python
df.Date
```

```python
type(df.Date[0])
# теперь мы видим что столбец Data имеет тип Timestamp
```

```python
# выведем индекс 
df.index
```

```python
# поле с датами не использовалось в качестве индекса для дата фрейма
# подправим програмный код с помощью параметра index_col функции pd.read_csv(), который указывает какой столбец в файле 
# следует использовать в качестве индекса

df = pd.read_csv('./goog.csv', 
                parse_dates = ['Date'],
                index_col = 'Date')
df
```

```python
# теперь индекс это DatetimeIndex
df.index
```
