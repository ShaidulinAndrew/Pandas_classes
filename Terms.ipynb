###  __В данном ноутбуке содержаться термины и определения используемые в анализе данных и библиотеке Pandas__

__Анализ данных__ - это процесс извлечения смысла из данных

__Анализ данных__ - это процесс получения информации из данных путем создания моделей и применения математического аппарата для поиска закономерностей.

__Наука о данных__ - это процесс использования статистики и анализа данных для понимания явлений, скрытых в данных.

__Аугментация данных__ - это создание дополнительных обучающих данных из имеющихся данных

__Структурированные данные__ - это тип данных, при котором последние упорядочены в вертикальные столбцы (поля) и горизонтальные строки (записи или наблюдения).

__Не структурированные данные__ - это данные, которые не имеют определенной структуры, не предполагают наличия заранее определенных столбцов определенного типа

__Полуструктурированные данные__ - это тип структурированных данных, у которых нет строгой структуры моделей данных

__Переменная__ - любая характеристика, число иил количество, которое можно измерить или сосчитать

__Категориальная переменная (categorical variable)__ - это переменная, которая может принимать одно значение из ограниченного и обычно фиксированного набора возможных значений.Категориальные переменные являются качественными характеристиками, которые могут описать продукт, но при этом не измеряется в неприрывной шкале.
__Categorical__ - специальный тип данных, который соответствует категориальным переменным в статистике.

__Непрерывная переменная (continuous variable)__ - это переменная которая может принимать бесконечное (неисчислимое) количество значений. Непрерывные переменные в библиотеке Pandas представлены либо типом float, либо типом integer.

__Временной ряд__ - представляет собой переменную, у которой значения зарегистрированы в определенные временные интервалы, таким образом, наблюдения изначально упорядочены во времени.

__Качественный анализ__ - это исследование данных, которые можно наблюдать, но нельзя количественно измерить

__Количественный анализ__ - это исследования данных, которые можно измерить количественно

__Одномерный анализ__ - это простейшая форма анализа данных и обычно используется для описания и подытоживания данных, а так же поиска закономерностей данных.

__Многомерный анализ__ - это метод моделирования в котором учавствуют две или более переменных, влияющих на результат эксперимента. Часто связан с такима понятиями как корелляция и регрессия, которые помогают понять взаимосвяь между несколькими переменными, а также влияние этой взаимосвязи на результат.

__Корреляция__ - это число, которое описывает степень взаимосвязи между двумя переменными.

__Регрессия__ - это статистический метод, который оценивает силу взаимосвязи между зависимой переменной и рядом других переменных.

__Series__ - базовая структура данных в библиотеке pandas. Series похож на массив в NumPy, но отличается от него тем, что имеет индекс, который позволяет осуществить гораздо более гибкий поиск элементов, а не только найти значение индекса массива, в котором отсчет начинается с нуля.

__Скаля́рная величинá__ (от лат. scalaris «ступенчатый») - величина, каждое значение которой может быть выражено одним (как правило, действительным) числом.

__Переиндексация в Pandas__ - это процесс, который согласовывает данные в серии с набором меток. Переиндексация выполняет несколько задач: переупорядочивание существующих данных в соответствии с набором меток; вставка маркеров Nan когда отсутствуют данные для метки; заполнение отсутствующих данных для метки с помощью тех или иных логических операций (по умолчанию используются значения Nan) 

__Принцип прямого заполнения__ - это метод последнего известного значения. Любым новым меткам индекса присваивается ранее известное значение, отличное от Nan
