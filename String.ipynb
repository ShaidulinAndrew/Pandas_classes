```python
import pandas as pd
import numpy as np
```

```python
# загружаем датафрейм 
data = pd.read_csv('./MFOcredit.csv', encoding = 'cp1251', sep = ';')
data.head()
```

### __Изменение регистра строк__

```python
# переводим строки (значения переменной marstatus) в нижний регистр
data['marstatus'] = data['marstatus'].str.lower()
data.head()
```

```python
# переводим строки (значения переменной marstatus) в верхний регистр
data['marstatus'] = data['marstatus'].str.upper()
data.head()
```

С помощью метода .str.replace() можно поменять одно строковое значение на другое

```python
# меняем одно из строковых значений переменной marstatus
data['marstatus'] = data['marstatus'].str.replace('ХОЛОСТ', 'Не женат/не замужем')
data.head()
```

### __Изменение строкового значения__

```python
# удаляем лишние символы из строковых значений переменных gender и jobtype
for i in ['gender', 'jobtype']:
    if i in data.columns:
        data[i] = data[i].str.replace('[*&_]', '')
data.head()
```

### __Определение пола клиента по отчеству__

```python
# загружаем CSV файл с ФИО клиентов по которым нужно определить пол
data3 = pd.read_csv('./Gender_based_on_middle_name.csv', encoding = 'cp1251', sep = ';')
data3.head(5)
```

```python
# создаем переменную пол, которая будет иметь значение True если строковое значение переменной клиент содержит паттерн
# -вна и False в противном случае
data3['Пол'] = data3['Клиент'].str.contains('вна')
data3.head(10)
```

```python
data3.info()
```

```python
# присвоим переменной Пол тип str 
data3['Пол'] = data3['Пол'].astype('str')
data3.info()
```

```python
# переименуем категории переменной Пол
d = {'False': 'Мужской', 'True': 'Женский'}
data3['Пол'] = data3['Пол'].map(d)
data3.head(10)
```

### __Удаление лишних символов из строк__

```python
# с помощью метода .str.lstrip() удалим ненужный символ подчеркивая, с которого начинается несколько значений переменной Клиент
# метод .str.lstrip() возвращает копию указанной строки, с начала (слева l - left) которой устранены указанные символы
data3['Клиент'] = data3['Клиент'].str.lstrip('_')
data3.head(20)
```

```python
# с помощью метода .str.rstrip() удалим лишние символы которыми заканчиваются некоторые значения переменной Возраст, 
# метод .str.rstrip() возвращает копию указанной строки, с конца (справа r - right) которой устранены указанные символы
data3['Возраст'] = data3['Возраст'].str.rstrip('&лет')
data3.head(20)
```

```python
# удаляем последние три символа в каждом строковом переменной Регион
data3['Регион'] = data3['Регион'].map(lambda x: str(x)[:-3])
data3.head(20)
```

### __Удаление повторяющихся строк__

```python
data3 = data3.drop_duplicates()
data3.head(15)
```

### __Извлечение нужных символов из строк__

```python
# загружаем данные
data4 = pd.read_csv('./Raw_text.csv', encoding = 'cp1251')
data4.head()
```

Видим что несколько переменных записано в один столбец


```python
# с помощью метода .str.extract() извлекаем даты из из столбца raw создав переменную date
data4['date'] = data4['raw'].str.extract('(....-..-..)', expand = True)
data4.head()
```

```python
# с помощью метода .str.extract() извлекаем цифры из столбца raw создав переменную gender
data4['gender'] = data4['raw'].str.extract('(\d)', expand = True)
data4.head()
```

```python
# с помощью метода .str.extract() извлекаем числа с плавающей точкой из столбца raw создав переменную score 
data4['score'] = data4['raw'].str.extract('(\d\d\d\d\.\d)', expand = True)
data4.head()
```

```python
# с помощью метода .str.extract() извлекаем текст из столбца raw создав переменную city  
data4['city'] = data4['raw'].str.extract('([A-Z]\w{0,})', expand = True)
data4.head()
```

```python
# удаляем raw
data4.drop(['raw'], axis = 1, inplace = True)
data4.head()
```
